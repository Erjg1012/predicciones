```python
import pandas as pd
import matplotlib.pyplot as plt

# Datos de la tabla
data = {
    "Empresa": [
        "Miller Lite", "Pepsi", "Stroh’s", "Fed’l Express", "Burger King", "Coca-Cola",
        "McDonald’s", "MCI", "Diet Cola", "Ford", "Levi’s", "Bud Lite", "ATT/Bell",
        "Calvin Klein", "Wendy’s", "Polaroid", "Shasta", "Meow Mix", "Oscar Meyer",
        "Crest", "Kibbles ‘N Bits"
    ],
    "Impactos (millones)": [
        32.1, 99.6, 11.7, 21.9, 60.8, 78.6, 92.4, 50.7, 21.4, 40.1, 40.8, 10.4, 88.9,
        12.0, 29.2, 38.0, 10.0, 12.3, 23.4, 71.1, 4.4
    ],
    "Gasto (millones USD, 1983)": [
        50.1, 74.1, 19.3, 22.9, 82.4, 40.1, 185.9, 26.9, 20.4, 166.2, 27.0, 45.6, 154.9,
        5.0, 49.7, 26.9, 5.7, 7.6, 9.2, 32.4, 6.1
    ]
}

df = pd.DataFrame(data)

# Gráfico
plt.figure(figsize=(10, 6))
plt.scatter(df["Gasto (millones USD, 1983)"], df["Impactos (millones)"], color='blue')
plt.title("Relación entre Gasto Publicitario y Número de Impactos (1983)")
plt.xlabel("Gasto Publicitario (millones de USD, 1983)")
plt.ylabel("Impactos (millones)")
plt.grid(True)

# Etiquetas de algunas empresas destacadas
for i, row in df.iterrows():
    if row["Gasto (millones USD, 1983)"] > 100 or row["Impactos (millones)"] > 90:
        plt.text(row["Gasto (millones USD, 1983)"], row["Impactos (millones)"], row["Empresa"], fontsize=8)

plt.tight_layout()
plt.show()
```
